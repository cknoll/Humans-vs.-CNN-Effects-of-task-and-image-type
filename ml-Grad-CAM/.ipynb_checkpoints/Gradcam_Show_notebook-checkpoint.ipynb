In this notebook we will run through the Gradcam-Method to locate Areas that were important for the Networks decision regarding a class.


```python
# First we import libraries needed.

## Standard libraries mostly used for data manipulation and plotting.
import os
import json
import math
import numpy as np 
import matplotlib.pyplot as plt
from PIL import Image
import pandas as pd
import re
import csv
import cv2

## PyTorch & torchvision
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.utils.data as data
import torch.optim as optim

import torchvision
from torchvision import transforms as transforms

## Custom function to create the models.
from model import construct_model

def get_top_labels(logits):
    return torch.topk(logits.flatten(),5).indices.cpu().numpy()
def get_top_probs(logits):
    return torch.max(logits.flatten(),5).indices.cpu().numpy()

transform_test = transforms.Compose(
    [transforms.Resize(224),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])])

# Create the labeldict from our label file
CATEGORIES_PATH = 'data/categories_places365.txt'

labeldict = {}
label_idx_dict = {}
with open(os.path.abspath(CATEGORIES_PATH), 'r') as read_file:
    lines = read_file.readlines()
    for line in lines:
        cleanline = line.strip().split(" ")
        labeldict[int(cleanline[1])] = cleanline[0]
        label_idx_dict[cleanline[0]] = int(cleanline[1])
        
```

```python
# Define Needed paths and parameters
# Path and name of model.
CHECKPOINT_PATH = "chris/ResNet152/Resnet152_1.pth"
modelname = "ResNet152"


#Since we aren't training we will use the cpu
device = torch.device("cpu") # if not torch.cuda.is_available() else torch.device("cuda:0")

#create the model and load it. 

model = construct_model(modelname)
model = model.to(device)

cpt = torch.load(CHECKPOINT_PATH,map_location=torch.device('cpu'))
model.load_state_dict(cpt['checkpoints'][1]['model_state_dict'])
```

```python
def count_parameters(model):
    return sum(p.numel() for p in model.parameters() if p.requires_grad)


count = count_parameters(model)
print(count)
```

We have loaded the model.

For this example a Resnet152 Architecture was used using so called Bottleneck Blocks.

At the end of the Resnet instead of some Fully Connected layers we have an average pooling, this is needed for the CAM method, which is a different method to find important regions in the image for certain classes. It is not necessary for the GradCAM method, but since we trained one Network to test different XAI-Methods on we use it in our Network-Architecture.

Let's look at the layers, we can see the different blocks used, throughout the network we decrease the size of each filter, while increasing the amount of them. Our 224x224x3 was transformed into a 16x16x2048 "image". The output can differ depending on your architecture.

```python
model.eval()
```

While Gradcam can use multiple layers the paper suggests that the final layer give the most information on which areas had a positive influence when deciding on a certain class.

Let's thus select the final layer as target layer.

```python
# layer4[-1] means the final block in the last layer. conv3 is the final convolutional filter within this block.
GRADCAM_TARGET_LAYER = model.layer4[-1].conv3
```

We can now takes some images and run the GradCAM algorithm on them. Let's look at our image first.

```python
image_1 = get_image('eyetracking/data/01_Bilder_Datensatz/lighthouse35.jpg')
image_2 = get_image('eyetracking/data/01_Bilder_Datensatz/windmill8.jpg')
image_3 = get_image('eyetracking/data/01_Bilder_Datensatz/desert20.jpg')
image_4 = get_image('eyetracking/data/01_Bilder_Datensatz/wheatfield48.jpg')
```

```python
# Networks train better on normalized data and every image needs to have the same size.
# We transform our images into an image tensor first, resize it and finally take a 224x244 crop.
# For our test images we have already made them quadratic, so cropping isn't needed and we just resize them.
# Note that in the worst case this might cut out our object from the image, but this has proofen to work really well in practice
image_1_t = transform_test(image_1)
image_2_t = transform_test(image_2)
image_3_t = transform_test(image_3)
image_4_t = transform_test(image_4)
```

```python
fig, axes = plt.subplots(1,3,constrained_layout = True)
fig.set_figheight(15)
fig.set_figwidth(15)
axes[0].imshow(image_1)
axes[0].set_title("Original")
axes[1].imshow(np.moveaxis(image_1_t.numpy(),0,2))
axes[1].set_title("Network input")
axes[2].imshow(normalize_image(np.moveaxis(image_1_t.numpy(),0,2)))
axes[2].set_title("Renormalized input as image")
```

```python
# Let's look at the network results for our first image.
# We use unsqueeze to create a batchsize of 1
out = model.eval()(image_1_t.unsqueeze(0))
# we use a softmax so our scores can be interpreted as a percentage of certainty.
out = F.softmax(out,dim=1)
#let's take our result tensor out of the network
logits = out.flatten().detach().cpu()
logits_np = out.flatten().detach().cpu().numpy()
# we can now use our certainties to find the labels with the higest certainty.
top_logits = get_top_labels(logits)
print(top_logits)
```

```python
## To make sense out of these labels let's use our labeldict

print("Our best label is " + labeldict[top_logits[0]] + " with " + str(logits_np[top_logits[0]]) + " certainty\n")

print("Our top 5 Labels are:\n")
for i in top_logits:
    print(str(i) + " - " + labeldict[i] + " with certainty: " + str(logits_np[i]))
```

When using a random crop the results might differ. In our case the lighthouse only got 8% certainty but at least made it into the top 5.

We can observe that many classes seem to have something to do with the sea, indicating that the boat and the water on the image might have a strong influence on the classification task which outweight the part with the lighthouse.

Let's see if we can find some support for this idea using the GradCAM algorithm

```python
# Let's import what we need for gradcam and define some functions for visualizations

from gradcam import get_gradcam

def mask_on_image(mask,img,alpha=0.5):
    heatmap = get_rgb_heatmap(mask)
    img = img.squeeze()
    cam_on_img = (1-alpha)*img + alpha*heatmap
    return np.copy(cam_on_img)

def get_rgb_heatmap(mask):
    heatmap = cv2.applyColorMap(np.uint8(255 * mask), cv2.COLORMAP_JET)
    heatmap = cv2.cvtColor(heatmap, cv2.COLOR_BGR2RGB)
    heatmap = np.float32(heatmap) / 255
    return np.copy(heatmap)

def get_cutoff_area(mask,img,cutoff = 0.5):
    for i in range(3):
        img[:,:,i] = np.where(mask>cutoff,img[:,:,i],0)
    return np.copy(img)


def get_percentage_of_image(mask,image,percentage):
    masked_image = np.zeros_like(image)
    n = mask.size
    sortedvalues = np.sort(mask.flatten('K'))[::-1]
    
    index = int(n/100*percentage)
    index_2 = n//100*percentage
    cutoff = sortedvalues[index]
    for i in range(3):
        masked_image[:,:,i] = np.where(mask-cutoff>0,image[:,:,i],0)
    return masked_image
```

```python
plot_data = []

#let's get the gradcam mask for the top 5 labels.
for i in top_logits:
    plot_data.append(get_gradcam(model,GRADCAM_TARGET_LAYER,image_1_t.unsqueeze(0),i))
```

```python
fig, axes = plt.subplots(2,6,constrained_layout = True)
fig.set_figheight(7)
fig.set_figwidth(15)
for i in range(5):
    axes[0][i+1].imshow(mask_on_image(plot_data[i][0],plot_data[i][1]))
    axes[0][i+1].set_title(labeldict[top_logits[i]])
    axes[1][i+1].imshow(get_percentage_of_image(plot_data[i][0],plot_data[i][1],percentage=5))
    axes[1][i+1].set_title(labeldict[top_logits[i]] + " - top 5%")
axes[0][0].imshow(plot_data[0][1])
axes[1][0].imshow(plot_data[0][1])
print(plot_data[i][0].shape)
```

As we theorized the boat and it's surrounding area are important in the classification for our different classes which all have something to do with the sea

Interestingly the lighthouse indeed has a stronger focus on the area of the lighthouse compared to the boat, but the boat and sea area in the foreground overwhelmed the network and it decided that different classes were more likely. We can tell by the percentages we saw before though that the network in general was unsure about the correct result

We can obvserve that even though we have very different labels, the most important gradcam area still remains with the windmill. For example one could have expected that for the "sky" laben we would at least see more of the sky in our gradcam mask, or for the ocean and beach the focus might be stronger towards the water area.

While there is less focus on the Windmill itself the Windmill is still given the highest importance, indicating that GradCam especially for classes with a lower likelyhood doesn't really give you class specific information, but rather defines areas that are of bigger importance when making the classification.

After all we would expect actually the opposite for other classes, since in the training process we would try to make other classes consider an area with a windmill in it as less likely for themselfs.

This does show that while GradCAM can be considered a powerfull tool in some regards it is not a perfect solution to finding an area that corresponds to a certain class in the images presented.
